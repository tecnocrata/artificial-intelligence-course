# Fraccion/Porcentaje de Stopwords en un corpus

```
# Esta funcion calculara el porcentaje de palabras stopwords en un texto
def stopwords_percentage(text):
  # Escogemos el diccionario en ingles
  stopwd = stopwords.words('english')
  # Sacamos las stopwords
  content_filtered =  [w for w in text if w.lower() not in stopwd]
  return len (content_filtered)/len(text)

stopwords_percentage(text1)
# El 58% del libro son stopwords
```

# Lexicons Enriquecidos (listas comparativas de palabras)
Este sera un nuevo recurso lexico, que se utilizara en la traduccion de palabras en diferentes idiomas.

```
# Importamos swadesh= una lista de palabras resistentes a los prestamos, en diferentes idiomas
from nltk.corpus import swadesh
# Todos los lenguajes de esas palabras
print (swadesh.fileids())
```

```
# Esas palabras en Espaniol
print (swadesh.words('es'))
# Esas palabras en Ingles
print (swadesh.words('en'))
```

```
# Este objeto me permitira hacer las traducciones de Frances a Espaniol
fr2es = swadesh.entries (['fr','es'])
# Es una lista de tuplas
print(fr2es)
```

```
# Creamos un diccionario de traduccion, con las palabras ya existentes en la lista
translate = dict(fr2es)
# Usamos el diccionario para traducir chien a espaniol
translate['chien']
```

# Wordnet & Synsets

```
# Es un recurso lexico que tuvo auge antes del deep-learning
# Es una base de datos de caracter lexico (en ingles)

# Descargamos el wordnet en espanol
nltk.download('omw')
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```
# Analizamos los synsets (grupo de sinonimos)
ss = wn.synsets('carro', lang='spa')
ss
```

```
# Analizamos el contenido especifico de los synsets
# Recorremos los synsets analizando su nombre y definicion
for syn in ss:
  print(syn.name(), ' : ', syn.definition())
  for n in syn.lemma_names():
    print('*', n)
```

# Hyponyms: (Hiponimos)
Son conceptos mas especificos, que la palabra raiz de la cual derivan

```
# Obteniendo los Hiponimos de car
print (ss[0].name())
print(ss[0].hyponyms())
```

# Hypernyms (Hiperonimos)
Son conceptos mas generales (arriba en el arbol de jerarquia) que la palabra analizada

```
# Obtenemos los Hiperonimos de car
print (ss[0].name())
print(ss[0].hypernyms())
```
